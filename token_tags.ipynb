# ***Đánh bio tags cho summary***

## ***Tạo hàm bio_tagging_single để thử nghiệm cho 1 vài sample***

```python
def bio_tagging_single(sample):
    summary = sample['summary']
    tokens = word_tokenize(summary, format="text").split()
    tags = ['O'] * len(tokens)

    def tag_entity(entity, label_prefix):
        if entity is None or not str(entity).strip():
            return
        entity_tokens = word_tokenize(entity, format="text").split()
        for i in range(len(tokens) - len(entity_tokens) + 1):
            if tokens[i:i+len(entity_tokens)] == entity_tokens:
                tags[i] = f'B-{label_prefix}'
                for j in range(1, len(entity_tokens)):
                    tags[i + j] = f'I-{label_prefix}'
                break

    tag_entity(sample.get('arg_subject'), 'ARG_SUBJECT')
    tag_entity(sample.get('trigger'), f'TRIGGER-{sample.get("event_type")}')
    tag_entity(sample.get('arg_location'), 'ARG_LOCATION')
    tag_entity(sample.get('arg_time'), 'ARG_TIME')

    return list(zip(tokens, tags))
```

## ***Tạo cột tokens cho df***

```python
# Thêm cột 'tokens' chứa kết quả word_tokenize từ summary
import numpy as np
def tokenize_summary(row):
    summary = row['summary']
    if pd.isna(summary) or not str(summary).strip():
        return []
    return word_tokenize(summary, format="text").split()

df['tokens'] = df.apply(tokenize_summary, axis=1)
# Hiển thị vài dòng đầu để kiểm tra
df[['summary', 'tokens']].head()
```

## ***Tạo cột bio_tags cho df***

```python
# === Hàm tạo BIO tags cải tiến ===
def create_bio_tags_improved(tokens, trigger, event_type):
    tags = ["O"] * len(tokens)
    if pd.isna(trigger) or pd.isna(event_type):
        return tags

    trigger = str(trigger).strip()
    if not trigger:
        return tags

    tokens_lower = [token.lower() for token in tokens]
    trigger_lower = trigger.lower()

    trigger_variants = [
        trigger.split(), trigger.split('_'), trigger.split('-'), [trigger]
    ]

    found = False
    for trig_tokens in trigger_variants:
        if not trig_tokens:
            continue
        trig_tokens_lower = [t.lower() for t in trig_tokens]
        for i in range(len(tokens) - len(trig_tokens) + 1):
            if tokens_lower[i:i+len(trig_tokens)] == trig_tokens_lower:
                tags[i] = f"B-{event_type}"
                for j in range(1, len(trig_tokens)):
                    tags[i + j] = f"I-{event_type}"
                found = True
                break
        if found:
            break

    if not found:
        trigger_words = re.findall(r'\b\w+\b', trigger_lower)
        for word in trigger_words:
            for i, token in enumerate(tokens_lower):
                if word in token or token in word:
                    if tags[i] == "O":
                        tags[i] = f"B-{event_type}"
                        break
            if found:
                break

    return tags

def create_arg_tags_improved(tokens, args_dict):
    tags = ["O"] * len(tokens)
    tokens_lower = [token.lower() for token in tokens]

    for role, value in args_dict.items():
        if pd.isna(value) or not str(value).strip():
            continue
        value = str(value).strip().lower()

        value_variants = [value.split(), value.split('_'), value.split('-'), [value]]
        found = False
        for val_tokens in value_variants:
            if not val_tokens:
                continue
            val_tokens_lower = [t.lower() for t in val_tokens]
            for i in range(len(tokens) - len(val_tokens) + 1):
                if tokens_lower[i:i+len(val_tokens)] == val_tokens_lower:
                    if all(tags[i+j] == "O" for j in range(len(val_tokens))):
                        tags[i] = f"B-Arg-{role}"
                        for j in range(1, len(val_tokens)):
                            tags[i + j] = f"I-Arg-{role}"
                        found = True
                        break
            if found:
                break

        if not found:
            value_words = re.findall(r'\b\w+\b', value)
            for word in value_words:
                for i, token in enumerate(tokens_lower):
                    if len(word) > 2 and (word in token or token in word):
                        if tags[i] == "O":
                            tags[i] = f"B-Arg-{role}"
                            break

    return tags
```

```python
import re

# Áp dụng hàm tạo BIO tags cải tiến cho DataFrame
df["bio_tags"] = df.apply(
    lambda row: create_bio_tags_improved(
        row["tokens"], 
        row.get("trigger", None), 
        row.get("event_type", None)
    ), axis=1
)

df["argument_tags"] = df.apply(
    lambda row: create_arg_tags_improved(
        row["tokens"], 
        {
            "Subject": row.get("arg_subject"),
            "Location": row.get("arg_location"),
            "Time": row.get("arg_time")
        }
    ), axis=1
)

df.head()
```
